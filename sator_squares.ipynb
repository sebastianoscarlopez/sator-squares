```python
import urllib.request

def get_words():
  url = "https://raw.githubusercontent.com/PolMrt/allwords/master/sp/words.txt"

  response = urllib.request.urlopen(url)
  return response.read().decode().split('\n')

def unique_words(words):
  return sorted(list(set(words)))

all_words = get_words()
all_words = unique_words(all_words)

# print(all_words[0:10])
```

```python
def replace_accents(word):
  accent_dict = {"á": "a", "é": "e", "í": "i", "ó": "o", "ú": "u"}
  new_word = ""
  for char in word:
      if char in accent_dict:
          new_word += accent_dict[char]
      else:
          new_word += char
  return new_word

unaccented_words = []
for word in all_words:
    unaccented_words.append(replace_accents(word.lower()))

# print(unaccented_words[0:10])
```

```python
square_size = 5

def get_correct_size_words():
  return [word for word in unaccented_words if len(word) == 5]

def invert_word(word):
  return word[::-1]

correct_size_words = get_correct_size_words()
available_words = [word for word in correct_size_words if invert_word(word) in correct_size_words]
palindrome_words = [word for word in available_words if invert_word(word) == word]
non_palindrome_words = [word for word in available_words if word not in palindrome_words]


print(palindrome_words)
print(non_palindrome_words)
```

```python
# TODO It only works with 5 letters
# TODO Is valid to use the same word multiple times?
# TODO Is valid to have more than one palindrome?
# TODO Is valid to use names?

def get_valid_first_words(central_word):
  acceptable_words = [word for word in available_words if word[2] == central_word[0]]
  acceptable_words_without_invert = []
  for word in acceptable_words:
    if invert_word(word) not in acceptable_words_without_invert:
      acceptable_words_without_invert.append(word)
  return acceptable_words_without_invert

def get_valid_second_words(central_word, first_word):
  acceptable_words = [word for word in available_words if word[2] == central_word[1] and word[0] == first_word[1] and word[-1] == first_word[-2]]
  return acceptable_words

def is_valid_third_word(central_word, first_word, second_word):
  third_word = first_word[1] + second_word[1] + central_word[1]+ second_word[-2] + first_word[-2]
  return third_word in available_words

sator_squares = []
for central_word in palindrome_words:
  # print(central_word)
  valid_first_words = get_valid_first_words(central_word)
  for first_word in valid_first_words:
    # print([central_word, first_word])
    second_words = get_valid_second_words(central_word, first_word)
    for second_word in second_words:
      if is_valid_third_word(central_word, first_word, second_word):
        sator_squares.append([first_word, second_word, central_word])

for sator_square in sator_squares:
  print(sator_square[0])
  print(sator_square[1])
  print(sator_square[2])
  print(sator_square[1][::-1])
  print(sator_square[0][::-1])
  print('-----')

print('total:' + str(len(sator_squares)))
```
